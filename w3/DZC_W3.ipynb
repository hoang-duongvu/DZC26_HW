```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
```

```python
spark = SparkSession.builder \
    .master("local[*]") \
    .appName('test') \
    .getOrCreate()
```

```python
df_large = spark.read.parquet("yellow_tripdata_2024-01.parquet")
for i in range(2, 7):
  url = f"yellow_tripdata_2024-{i:02d}.parquet"
  df_small = spark.read.parquet(url)
  df_large = df_large.union(df_small)

df_large.printSchema()
```

```python
df_large.count()
```

```python
df_fareamount = df_large.filter(col("fare_amount") == 0)
df_fareamount.count()
```

```python
df_pulocation = df_large.select("PULocationID").distinct()
df_pulocation.show()
```
